```python
import os
import shutil
```

```python
# --- Configuration ---
# 1. The original file with "path text" lines.
# input_file = 'train.txt'
input_file = 'val.txt'

# 2. The folder where you want to move the images.
# dest_folder = 'train_images'
dest_folder = 'val_images'

# 3. The new file that will store the updated paths.
# output_file = 'train_refined.txt'
output_file = 'val_refined.txt'
# --- End Configuration ---
```

```python
# Create the destination folder if it doesn't exist
os.makedirs(dest_folder, exist_ok=True)

print(f"Reading from '{input_file}'...")
print(f"Copied images to '{dest_folder}'...")

# Open both files
try:
    with open(input_file, 'r') as f_in, open(output_file, 'w') as f_out:
        # Read each line from the input file
        for line in f_in:
            line = line.strip()
            if not line:
                continue

            # Split the line into path and the rest of the text
            try:
                original_path, text_content = line.split(' ', 1)
            except ValueError:
                print(f"Warning: Skipping malformed line: '{line}'")
                continue

            # original_path = os.path.join('line_images', original_path)

            # Check if the source image file actually exists
            if os.path.exists(original_path):
                # Get just the filename from the original path
                filename = os.path.basename(original_path)
                
                # Create the new path in the destination folder
                new_path = os.path.join(dest_folder, filename)

                # Move the file
                shutil.copy2(original_path, new_path)

                # Write the new path and the text to the output file
                f_out.write(f"{new_path} {text_content}\n")
                
                print(f"Copied '{original_path}' -> '{new_path}'")
            else:
                print(f"Warning: File not found, skipping: '{original_path}'")

    print(f"\nDone! Updated paths are saved in '{output_file}'.")

except FileNotFoundError:
    print(f"Error: The input file '{input_file}' was not found.")
```
