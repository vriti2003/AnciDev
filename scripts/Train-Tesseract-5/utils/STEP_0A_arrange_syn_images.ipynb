```python
import os
import shutil
```

```python
# Define the source and destination directories
source_dir = '../images'
output_dir = '../syn_images'
```

```python
# The exist_ok=True part prevents an error if the folder is already there
os.makedirs(output_dir, exist_ok=True)

# Loop through each subfolder in the source directory (e.g., 'Akshar-Bold')
for subfolder_name in os.listdir(source_dir):
    subfolder_path = os.path.join(source_dir, subfolder_name)

    # Make sure it's actually a folder
    if os.path.isdir(subfolder_path):
        # Loop through each file inside the subfolder
        for filename in os.listdir(subfolder_path):
            # Create the full path to the original file
            source_file_path = os.path.join(subfolder_path, filename)

            # Create the new filename by joining the folder and file name
            new_filename = f"{subfolder_name}_{filename}"

            # Create the full path for the new destination file
            destination_file_path = os.path.join(output_dir, new_filename)

            # Copy the file to the new location with the new name
            shutil.copy2(source_file_path, destination_file_path)
            print(f"Copied: {filename} -> {new_filename}")

print("\nAll files have been reorganized successfully! ✨")
```
