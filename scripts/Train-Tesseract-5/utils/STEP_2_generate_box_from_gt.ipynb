```python
img_dir = '../val_images'
gt_dir = '../val_texts'
box_dir = '../val_images'

os.makedirs(box_dir, exist_ok=True)
```

```python
for fname in os.listdir(gt_dir):
    if fname.endswith('.txt'):
        basename = os.path.splitext(fname)[0]
        txt_path = os.path.join(gt_dir, fname)
        img_path = os.path.join(img_dir, f"{basename}.jpg")

        if not os.path.exists(img_path):
            print(f"Image not found for: {basename}")
            continue

        # Read text
        with open(txt_path, 'r', encoding='utf-8') as f:
            text = f.read().strip()

        # Get image size
        im = Image.open(img_path)
        w, h = im.size

        # Write bounding box file
        with open(os.path.join(box_dir, f"{basename}.box"), 'w', encoding='utf-8') as f:
            for ch in text:
                f.write(f"{ch} 0 0 {w} {h} 0\n")
```
