```python
import os
```

```python
def split_text_by_image(input_file, output_dir="output_texts"):
    # Create output folder if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    with open(input_file, "r", encoding="utf-8") as infile:
        for line in infile:
            line = line.strip()
            if not line:
                continue

            # Split only at the first space
            parts = line.split(" ", 1)
            if len(parts) < 2:
                continue

            image_name, text = parts
            base_name = str(os.path.splitext(image_name)[0])
            base_name = base_name.split("/")[1]
            print(base_name)

            # Create output text file
            output_path = os.path.join(output_dir, f"{base_name}.txt")
            with open(output_path, "w", encoding="utf-8") as outfile:
                outfile.write(text)

    print(f"✅ Text files created in folder: {output_dir}")
```

```python
# Example usage
# split_text_by_image("./train_refined.txt", "train_texts")
split_text_by_image("./val_refined.txt", "val_texts")
```
