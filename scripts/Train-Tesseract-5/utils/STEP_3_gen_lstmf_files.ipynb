```python
img_dir = '../val_images'
box_dir = '../val_images'

lstmf_dir = '../val_lstmf'
os.makedirs(lstmf_dir, exist_ok=True)
```

```python
for fname in os.listdir(img_dir):
    if fname.endswith('.jpg'):
        basename = os.path.splitext(fname)[0]
        img_path = os.path.join(img_dir, fname)
        box_path = os.path.join(box_dir, f"{basename}.box")
        output_path = os.path.join(lstmf_dir, f"{basename}.lstmf")

        # print(basename)
        # print(img_path)
        # print(box_path)
        # print(output_path)

        # print(f'img path : {img_path}')
        # print(f'box_path path : {box_path}')
        # print(f'output_path path : {output_path}')
        # cmd = f'tesseract {img_path} {output_path[:-6]} -l hin lstm.train --boxes {box_path}'
        cmd = f'tesseract {img_path} {output_path[:-6]} -l hin lstm.train'
        try:
            subprocess.run(cmd, shell=True)
        except e:
            continue
```
