# Merge Text Files with Annotations

This notebook reads all `.txt` files from a specified folder, extracts the annotation from the second line (skipping the first blank line), and merges them into a single output file with the format:

`<file_path> <annotation>`

```python
import os
from pathlib import Path
import glob
```

```python
# Configuration
input_folder = "op_text_files"  # Change this to your folder path
output_file = "merged_annotations.txt"    # Output file name
```

```python
# Debug: Check if folder exists and list its contents
print(f"Checking folder: {input_folder}")
print(f"Absolute path: {os.path.abspath(input_folder)}")
print(f"Folder exists: {os.path.exists(input_folder)}")
print(f"Is directory: {os.path.isdir(input_folder)}")

if os.path.exists(input_folder):
    print(f"\nAll files in folder:")
    all_files = os.listdir(input_folder)
    for f in all_files[:20]:  # Show first 20 files
        print(f"  - {f}")
    if len(all_files) > 20:
        print(f"  ... and {len(all_files) - 20} more files")
    
    # Count .txt files
    txt_files = [f for f in all_files if f.endswith('.txt')]
    print(f"\n.txt files found: {len(txt_files)}")
```

```python
def merge_txt_files(input_folder, output_file):
    """
    Merge all .txt files from input_folder into a single output file.
    
    Args:
        input_folder (str): Path to folder containing .txt files
        output_file (str): Path to output merged file
    """
    # Method 1: Using glob (more flexible with path formats)
    pattern = os.path.join(input_folder, '*.txt')
    txt_files = sorted(glob.glob(pattern))
    
    # Alternative Method 2: Using os.listdir if glob doesn't work
    if not txt_files:
        print("Trying alternative method...")
        if os.path.exists(input_folder):
            all_files = os.listdir(input_folder)
            txt_files = sorted([os.path.join(input_folder, f) for f in all_files if f.endswith('.txt')])
    
    if not txt_files:
        print(f"No .txt files found in {input_folder}")
        print(f"Please check the path and try again.")
        return
    
    print(f"Found {len(txt_files)} .txt files")
    print(f"\nProcessing files...")
    
    # Open output file for writing
    processed_count = 0
    with open(output_file, 'w', encoding='utf-8') as out_f:
        for txt_file in txt_files:
            try:
                with open(txt_file, 'r', encoding='utf-8') as in_f:
                    lines = in_f.readlines()
                    
                    # Check if file has at least 2 lines
                    if len(lines) >= 2:
                        # Get the second line (index 1) and strip whitespace
                        annotation = lines[1].strip()
                        
                        # Write file path and annotation to output
                        out_f.write(f"{txt_file} {annotation}\n")
                        processed_count += 1
                    else:
                        print(f"Warning: {txt_file} has less than 2 lines, skipping...")
                        
            except Exception as e:
                print(f"Error processing {txt_file}: {e}")
    
    print(f"\nSuccessfully processed: {processed_count} files")
    print(f"Merged annotations saved to: {output_file}")
```

```python
# Run the merge operation
merge_txt_files(input_folder, output_file)
```

```python
# Preview the merged file
if os.path.exists(output_file):
    print("Preview of merged file (first 10 lines):")
    print("=" * 80)
    with open(output_file, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f):
            if i >= 10:
                break
            print(line.rstrip())
else:
    print(f"Output file {output_file} not found.")
```

```python
# Count total lines in output
if os.path.exists(output_file):
    with open(output_file, 'r', encoding='utf-8') as f:
        total_lines = sum(1 for _ in f)
    print(f"\nTotal annotations merged: {total_lines}")
else:
    print(f"Output file {output_file} not found.")
```
