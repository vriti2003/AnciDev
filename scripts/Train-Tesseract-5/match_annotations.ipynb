# Match Annotations from Two Files

This notebook matches annotations from two files based on image names:
- **merged_annotations.txt**: `op_text_files/image_name.txt annotation`
- **val.txt**: `line_images/image_name.jpg annotation`

Output format: `merged_annotation****val_annotation`

```python
import os
from pathlib import Path
```

```python
# Configuration
merged_annotations_file = "merged_annotations.txt"
val_file = "val.txt"
output_file = "matched_annotations.txt"
```

```python
def extract_image_name(filepath):
    """
    Extract the base image name without extension from a file path.
    
    Args:
        filepath: Full path like 'op_text_files/image_name.txt' or 'line_images/image_name.jpg'
    
    Returns:
        Base name without extension (e.g., 'image_name')
    """
    # Get the filename from the path
    filename = os.path.basename(filepath)
    # Remove extension
    base_name = os.path.splitext(filename)[0]
    return base_name
```

```python
# Test the extract function
print("Testing extract_image_name function:")
print(f"  'op_text_files/image001.txt' -> '{extract_image_name('op_text_files/image001.txt')}'")
print(f"  'line_images/image001.jpg' -> '{extract_image_name('line_images/image001.jpg')}'")
```

```python
# Read merged_annotations.txt and create a dictionary
merged_dict = {}

print(f"Reading {merged_annotations_file}...")
with open(merged_annotations_file, 'r', encoding='utf-8') as f:
    for line_num, line in enumerate(f, 1):
        line = line.strip()
        if not line:
            continue
        
        # Split by first space to separate path and annotation
        parts = line.split(' ', 1)
        if len(parts) >= 2:
            filepath = parts[0]
            annotation = parts[1]
            
            # Extract image name
            image_name = extract_image_name(filepath)
            merged_dict[image_name] = annotation
        else:
            print(f"Warning: Line {line_num} doesn't have proper format")

print(f"Loaded {len(merged_dict)} entries from merged_annotations.txt")
print(f"\nFirst 5 entries:")
for i, (key, value) in enumerate(list(merged_dict.items())[:5]):
    print(f"  {key}: {value[:50]}..." if len(value) > 50 else f"  {key}: {value}")
```

```python
# Read val.txt and match with merged_dict
matched_count = 0
unmatched_count = 0
unmatched_names = []

print(f"Reading {val_file} and matching...\n")
with open(val_file, 'r', encoding='utf-8') as f_val, \
     open(output_file, 'w', encoding='utf-8') as f_out:
    
    for line_num, line in enumerate(f_val, 1):
        line = line.strip()
        if not line:
            continue
        
        # Split by first space to separate path and annotation
        parts = line.split(' ', 1)
        if len(parts) >= 2:
            filepath = parts[0]
            val_annotation = parts[1]
            
            # Extract image name
            image_name = extract_image_name(filepath)
            
            # Check if this image exists in merged_dict
            if image_name in merged_dict:
                merged_annotation = merged_dict[image_name]
                
                # Write to output: merged_annotation****val_annotation
                f_out.write(f"{merged_annotation}****{val_annotation}\n")
                matched_count += 1
            else:
                unmatched_names.append(image_name)
                unmatched_count += 1
        else:
            print(f"Warning: Line {line_num} in {val_file} doesn't have proper format")

print(f"{'='*60}")
print(f"Results:")
print(f"  Matched entries: {matched_count}")
print(f"  Unmatched entries: {unmatched_count}")
print(f"  Output saved to: {output_file}")
print(f"{'='*60}")

if unmatched_count > 0 and unmatched_count <= 10:
    print(f"\nUnmatched image names:")
    for name in unmatched_names:
        print(f"  - {name}")
elif unmatched_count > 10:
    print(f"\nFirst 10 unmatched image names:")
    for name in unmatched_names[:10]:
        print(f"  - {name}")
```

```python
# Preview the output file
if os.path.exists(output_file):
    print(f"Preview of {output_file} (first 10 lines):")
    print("="*80)
    with open(output_file, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f, 1):
            if i > 10:
                break
            # Show first 100 chars of each line
            line_preview = line.rstrip()
            if len(line_preview) > 100:
                line_preview = line_preview[:100] + "..."
            print(f"{i}: {line_preview}")
else:
    print(f"Output file {output_file} not found.")
```

```python
# Count total lines in output
if os.path.exists(output_file):
    with open(output_file, 'r', encoding='utf-8') as f:
        total_lines = sum(1 for _ in f)
    print(f"\nTotal matched annotations in output: {total_lines}")
else:
    print(f"Output file {output_file} not found.")
```
