```python
#!/usr/bin/env python3
"""
Script to create train.txt and val.txt with 80:20 split from line annotations.
Prepends 'line_images/' to each entry.
"""

import os
from pathlib import Path
import random
from typing import List

def find_line_annotations(base_path: str) -> List[Path]:
    """Find all annotations.txt files in the directory structure."""
    base = Path(base_path)
    annotation_files = list(base.rglob("line_images/annotations.txt"))
    return annotation_files

def collect_all_entries(annotation_files: List[Path]) -> List[str]:
    """Collect all entries from annotation files."""
    all_entries = []
    
    for anno_file in annotation_files:
        print(f"Processing: {anno_file}")
        try:
            with open(anno_file, 'r', encoding='utf-8') as f:
                lines = f.readlines()
                # Strip whitespace and filter empty lines
                lines = [line.strip() for line in lines if line.strip()]
                all_entries.extend(lines)
        except Exception as e:
            print(f"Error reading {anno_file}: {e}")
    
    return all_entries

def create_train_val_split(entries: List[str], train_ratio: float = 0.8, 
                          output_dir: str = '.', seed: int = 42):
    """
    Create train.txt and val.txt with specified split ratio.
    Prepends 'line_images/' to each entry.
    """
    # Shuffle entries for random split
    random.seed(seed)
    shuffled_entries = entries.copy()
    random.shuffle(shuffled_entries)
    
    # Calculate split index
    split_idx = int(len(shuffled_entries) * train_ratio)
    
    train_entries = shuffled_entries[:split_idx]
    val_entries = shuffled_entries[split_idx:]
    
    print(f"\nTotal entries: {len(shuffled_entries)}")
    print(f"Train entries: {len(train_entries)} ({len(train_entries)/len(shuffled_entries)*100:.1f}%)")
    print(f"Val entries: {len(val_entries)} ({len(val_entries)/len(shuffled_entries)*100:.1f}%)")
    
    # Prepend 'line_images/' and write train.txt
    train_path = os.path.join(output_dir, 'train.txt')
    with open(train_path, 'w', encoding='utf-8') as f:
        for entry in train_entries:
            f.write(f"line_images/{entry}\n")
    print(f"\nCreated: {train_path}")
    
    # Prepend 'line_images/' and write val.txt
    val_path = os.path.join(output_dir, 'val.txt')
    with open(val_path, 'w', encoding='utf-8') as f:
        for entry in val_entries:
            f.write(f"line_images/{entry}\n")
    print(f"Created: {val_path}")

def main():
    # Configuration
    base_path = "../manuscript_dataset/Archive_Org"
    output_dir = "."
    train_ratio = 0.8
    random_seed = 42
    
    # Check if base path exists
    if not os.path.exists(base_path):
        print(f"Error: Base path '{base_path}' does not exist!")
        print(f"Current directory: {os.getcwd()}")
        print("\nPlease update the 'base_path' variable in the script to match your directory structure.")
        return
    
    print(f"Searching for annotations.txt files in: {base_path}\n")
    
    # Find all annotation files
    annotation_files = find_line_annotations(base_path)
    
    if not annotation_files:
        print("No line_annotations.txt files found!")
        print("Expected path structure: */line_images/annotations.txt")
        return
    
    print(f"Found {len(annotation_files)} annotation files\n")
    
    # Collect all entries
    all_entries = collect_all_entries(annotation_files)
    
    if not all_entries:
        print("No entries found in annotation files!")
        return
    
    # Create train/val split
    create_train_val_split(all_entries, train_ratio, output_dir, random_seed)
    
    print("\n✓ Done!")

if __name__ == "__main__":
    main()
```
